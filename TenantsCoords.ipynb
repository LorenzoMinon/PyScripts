```python
import pandas as pd
import requests

# Tu API Key de HERE
API_KEY = "insert"

# Endpoint para geocodificación
endpoint = "insert"

# Cargar el archivo Excel
file_path = "TenantsFiltered50.xlsx"  # Reemplaza con la ruta de tu archivo
data = pd.read_excel(file_path)

# Columns!
data['full_address'] = data.apply(
    lambda row: f"{row['address1']}, {row['city']}, {row['state']}, {row['postal']}, {row['country']}", axis=1
)

# Función para geocodificar con la API de HERE
def geocode_address(address):
    params = {
        "q": address,
        "apiKey": API_KEY
    }
    try:
        response = requests.get(endpoint, params=params)
        if response.status_code == 200:
            data = response.json()
            if "items" in data and len(data["items"]) > 0:
                location = data["items"][0]
                return location["position"]["lat"], location["position"]["lng"]
        return None, None  # Si no encuentra resultados
    except Exception as e:
        print(f"Error al procesar la dirección '{address}': {e}")
        return None, None  # Si ocurre algún error

# Aplicar la geocodificación a todas las filas
data[['latitude', 'longitude']] = data['full_address'].apply(
    lambda addr: pd.Series(geocode_address(addr))
)

# Guardar el archivo actualizado con las columnas completadas
output_path = "TenantsUpdatedPart1.xlsx"  # Reemplaza con la ruta donde deseas guardar el archivo
#data.to_excel(output_path, index=False)

print(f"Archivo actualizado con latitud y longitud guardado en: {output_path}")
```
